```python
import numpy as np
import pandas as pd

# give you file path here
dataset = pd.read_csv('/content/Bank Customer Churn Prediction.csv')

dataset.head()
```

```python
print(dataset.isnull().sum())
print(dataset.describe())
```

```python
dataset.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count the values in the churn column
print(dataset['churn'].value_counts())

# Plot the churn distribution
sns.countplot(x='churn', data=dataset)
plt.title('Churn Distribution')
plt.xlabel('Churn (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
dataset['estimated_salary'] = pd.to_numeric(dataset['estimated_salary'], errors='coerce')
dataset['estimated_salary'] = dataset['estimated_salary'].fillna(dataset['estimated_salary'].median())
```

```python
dataset.info()
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()

categorical_cols = ['country', 'gender']

for col in categorical_cols:
    dataset[col] = labelencoder.fit_transform(dataset[col])
```

```python
from sklearn.model_selection import train_test_split

# Features (X) and Target (y)
X = dataset.drop(['customer_id', 'churn'], axis=1)
y = dataset['churn']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    random_state=0
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")
```
